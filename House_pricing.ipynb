```
x = df.copy()
y= x.pop("Price")
#x = df.iloc[:, :-1].values
#y = df.iloc[:, 9].values
print(x.shape)
#print(x)
#print(y)

for ColName in x[['Bathroom','Furnishing','Parking','Type','Transaction']]:
  dummy = pd.get_dummies(x[ColName])
  x = pd.concat([x,dummy], axis  = 1)
  
print(x.head(3))
print(x.shape)

for ColName in x[['Bathroom','Furnishing','Parking','Type','Transaction']]:
  x.drop([ColName], axis = 1,inplace = True) 

print(x.head(3))
print(x.shape)
```

```

top_50 = [ i for i in df.Locality.value_counts().sort_values(ascending = False).head(50).index]
#top_50
```

```
for label in top_50:
  x[label] = np.where(x['Locality'] == label ,1,0)

x.drop(['Locality'], axis = 1,inplace = True) 
x
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y , test_size = 0.2, random_state=0)
```

```
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)

y_pred = regressor.predict(x_test)

from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(r2*100,"%")
```
